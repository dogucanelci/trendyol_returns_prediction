```python
df_user_demog.birth_date.sort_values() #There is some unusual dates which avoid convert to datetime format
df_user_demog_nan = df_user_demog.dropna(subset=['birth_date'])
df_user_demog_nan
```

```python
#Remove rows that has outlier datetimes
df_user_demog_nan.birth_date = pd.to_datetime(df_user_demog_nan.birth_date,errors='coerce')
df_user_demog_nan.membership_date = pd.to_datetime(df_user_demog_nan.membership_date,errors='coerce')
df_user_demog_nan['birth_year'] = df_user_demog_nan.birth_date.dt.year
df_user_demog_nan =df_user_demog_nan[(df_user_demog_nan.birth_year > 1900) & (df_user_demog_nan.birth_year < 2019)]
df_user_demog_nan
```

```python
#user_member_date_age : age of the customer at the time he or she became a member
df_user_demog_nan['membership_date'] = pd.to_datetime(df_user_demog_nan['membership_date'])
df_user_demog_nan['user_member_date_age'] = df_user_demog_nan.membership_date.dt.year -  df_user_demog_nan.birth_year
df_user_demog_nan
```

```python
#remove hour and seconds from membership_date
df_user_demog_nan['membership_date1'] = df_user_demog_nan['membership_date'].dt.date
```

## Encoding Part

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_user_demog_nan['gender'] = encoder.fit_transform(df_user_demog_nan['gender'])
df_user_demog_nan
```

# Transactions for getting User return rates

```python
df_trans = pd.read_csv('./datas/transactions.csv')
df_trans.head()
```

```python
df_trans_train = df_trans.dropna(subset=['is_returned'])
df_trans_train.is_returned.isna().sum()
```

```python
#users return rate per basket
```

```python
# Number of total returned items per order_packet and its user
df_trans_grouped = df_trans_train.groupby(['user_id','order_parent_id'])['is_returned'].sum()
```

```python
df_trans_grouped1 = pd.DataFrame(df_trans_grouped).reset_index()
```

```python
df_trans_grouped1['total_basket_count'] = 0
df_trans_grouped1['returned_basket_count'] = 0
df_trans_grouped1['basket_returned_rate'] = 0
#total_basket_count: number of order_packet per user has since today
df_trans_grouped1['total_basket_count'] = df_trans_grouped1.groupby('user_id').transform('count')
#returned_basket_count: number of returned order_packet per user has since today 
df_trans_grouped1['returned_basket_count'] = df_trans_grouped1[df_trans_grouped1.is_returned > 0].groupby('user_id').transform('count')
df_trans_grouped1.fillna(0,inplace=True)
#returned_basket_count: ratio of number of returned order_packet to number of all order packet
df_trans_grouped1['basket_returned_rate'] = df_trans_grouped1['returned_basket_count'] / df_trans_grouped1['total_basket_count']
```

```python
#users last elite member status

#Some users may decided to be an elite member after a while or cancel it, so it must be determined to
#last status of the user.
```

```python
df_trans_train.order_date = pd.to_datetime(df_trans_train.order_date)
```

```python
# Last order dates of all users
df_trans_train.groupby(['user_id'])['order_date'].max()
```

```python
# All order rows which has last orders of all users are fetched.
df_last_elite = df_trans_train.sort_values(by='order_date',ascending=False).groupby('user_id').first().reset_index()
df_last_elite = df_last_elite.loc[:,['user_id','order_date','is_elite_user']]
```

```python
# Last elite user or not status of all users are pushed into the column.
df_users_main_1 = pd.merge(df_trans_grouped1,df_last_elite,on=['user_id'],how='left').drop_duplicates(subset=['user_id'])
df_users_main_1.head()
```

```python
# All features created before are merged at one dataframe.
df_users_main_2 = pd.merge(df_users_main_1,df_user_demog_nan,on=['user_id'],how='left')
del(df_users_main_1)
df_users_main_2.head()
```

```python
# Some historic columns are added the main user dataframe.
df_users_main_2['last_order_date'] = pd.to_datetime(pd.to_datetime(df_users_main_2.order_date).dt.date)
df_users_main_2['membership_date1'] = pd.to_datetime(df_users_main_2.membership_date1)
```

```python
df_users_main_2.head()
```

```python
# current_member_time: Total member time of all users.
df_users_main_2['current_member_time'] = df_users_main_2.last_order_date.dt.year - df_users_main_2.membership_date1.dt.year
df_users_main_2.head()
```

```python
# Columns used for model training are filtered.
df_users_main_2 = df_users_main_2.loc[:,['user_id','returned_basket_count','basket_returned_rate', 'total_basket_count','is_elite_user', 'gender', 'user_member_date_age', 'current_member_time']]
df_users_main_2.head()
```
