```python
#Option-1 : Preprocessing with Turkish words Stemmizer
def preprocess_func_with_turkish_stemmer(name):
    # Instantiate stemming class
    #stemmer = PorterStemmer()
    stemmer = TurkishStemmer()
    #Import the english stop words list from NLTK
    stopwords_english = stopwords.words('turkish')
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    # tokenize names
    name_tokens = tokenizer.tokenize(name)
    names_clean = []
    for word in name_tokens:
        if (word not in stopwords_english and  # remove stopwords
                word not in string.punctuation):  # remove punctuation
            # names_clean.append(word)
            stem_word = stemmer.stem(word)  # stemming word
            if re.findall(r'\w*\d\w*',stem_word) == []:
                names_clean.append(stem_word)
    return names_clean
```

```python
#Option-2 : Preprocessing without any Stemmizer
def preprocess_func(name):

    stopwords_english = stopwords.words('turkish')
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    # tokenize names
    name_tokens = tokenizer.tokenize(name)
    names_clean = []
    for word in name_tokens:
        if (word not in stopwords_english and  # remove stopwords
                word not in string.punctuation):  # remove punctuation
            # names_clean.append(word)
            #stem_word = stemmer.stem(word)  # stemming word
            if re.findall(r'\w*\d\w*',word) == []:
                names_clean.append(word)
    return names_clean
```

```python
def get_corpus(df,column_name):
    #Some of product names are gone when preprocessing is applied, so nan values are set as '-1'.
    df[column_name].fillna('-1',inplace=True)
    #preprocessed product names are converted to token list for creating a corpus.
    corpus = [tokens.split(' ') for tokens in df[column_name].values]
    return corpus
```

```python
#get product name corpus for training of word2vec model
prod_name_corpus = get_corpus(df_prods,'product_name_prep')
```

```python
len(prod_name_corpus)
```

```python
#In this NLP problem which is actually representation of product names, word2vec model is selected. It means that co occurance of tokens
#takes first importance.
#Other ways might be more efficient.(FastText,GloVe,Bert or embedding layer etc.)
model_product_name = word2vec.Word2Vec(prod_name_corpus, vector_size=20, window=3, min_count=2, workers=4)
```

<strong>Word2vec</strong> model is trained. So every token in the corpus have 20 dim. numerical representation.

```python
#It convert product name tokens to vector and then get mean of all token vector of product name to
#get representation of all product name.
def add_word2vec_column(x,model,emded_size):
    temp_list = []
    #df_temp = x.product_name_prep
    for token in x:
        if token in model.wv.key_to_index:
                wordvec = model.wv.get_vector(token)
                temp_list.append(wordvec)
        else:
            temp_list.append([0 for x in range(emded_size)])
    product_name_vector = np.mean(temp_list,axis=0)
    return product_name_vector
```

```python
df_prods['embed_prod_name'] = df_prods.apply(lambda x: add_word2vec_column(x.product_name_prep,model_product_name,20),axis=1)
```
